```python
data=pd.read_csv("creditcard.csv")
data.head()
```

```python
plt.figure(figsize=(25,15)) 
sns.heatmap(data.corr(), annot=True, linewidths=.3, fmt= '.2f',cmap='viridis')
plt.show()
```

**Handling Imbalance Dataset**

```python
sns.countplot(x=data.iloc[:,-1],data=data,palette='rainbow')
plt.show()
```

```python
X=data.iloc[:,:-1]
y=data['Class']
```

```python
y.value_counts()
```

```python
#oversampling
from imblearn.over_sampling import SMOTE #create synthetic/artificial points
oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
```

```python
y.value_counts() #balance
```

```python
sns.countplot(x=y,data=data,palette='rainbow')
plt.show()
```

**Training the Model**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=64)
```

```python
from sklearn.svm import SVC
model=SVC(kernel='linear')
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
Y_pred1=model.predict(X_train)
```

**Performance**

```python
#testing accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#training accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_train,Y_pred1)
```

```python
# testing
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
